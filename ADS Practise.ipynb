```python
# Find Maximum and Minimum in 1D Array

# Input number of elements
n = int(input("Enter number of elements: "))

# Input array elements
arr = []
print("Enter", n, "elements:")
for i in range(n):
    value = int(input())
    arr.append(value)

# Initialize max and min with the first element
maximum = arr[0]
minimum = arr[0]

# Traverse array to update max and min
for i in range(1, n):
    if arr[i] > maximum:
        maximum = arr[i]
    if arr[i] < minimum:
        minimum = arr[i]

# Display result
print("Maximum element:", maximum)
print("Minimum element:", minimum)
```

```python
# Sorting an array in ascending order using Bubble Sort

# Input array
arr = list(map(int, input("Enter array elements: ").split()))
n = len(arr)

# Bubble Sort
for i in range(n):
    for j in range(0, n - i - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Swap

# Display result
print("Sorted array in ascending order:", arr)
```

```python
arr = [1, 2, 3, 4, 5]
left = 0
right = len(arr) - 1

while left < right:
    arr[left], arr[right] = arr[right], arr[left]
    left += 1
    right -= 1

print("Reversed Array:", arr)
```

```python
# Reverse the array without using reverse()

# Input array
arr = list(map(int, input("Enter array elements: ").split()))
n = len(arr)

# Two-pointer technique
start = 0
end = n - 1

while start < end:
    arr[start], arr[end] = arr[end], arr[start]  # Swap
    start += 1
    end -= 1

# Display reversed array
print("Reversed array:", arr)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def push_front(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        print(f"{data} added at front.")

    def push_back(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        print(f"{data} added at back.")

    def delete_value(self, value):
        if not self.head:
            print("List is empty.")
            return

        # delete head node
        if self.head.data == value:
            self.head = self.head.next
            if not self.head:
                self.tail = None
            print(f"{value} deleted.")
            return

        prev = None
        curr = self.head
        while curr:
            if curr.data == value:
                prev.next = curr.next
                if curr == self.tail:
                    self.tail = prev
                print(f"{value} deleted.")
                return
            prev = curr
            curr = curr.next

        print(f"{value} not found in list.")

    def print_list(self):
        if not self.head:
            print("List is empty.")
            return
        curr = self.head
        print("Current List:", end=" ")
        while curr:
            print(curr.data, end=" ")
            curr = curr.next
        print("\n")


# -------- Main Menu --------
ll = SinglyLinkedList()

while True:
    print("\n--- Singly Linked List Menu ---")
    print("1. Push Front")
    print("2. Push Back")
    print("3. Delete Value")
    print("4. Print List")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        val = int(input("Enter value to add at front: "))
        ll.push_front(val)

    elif choice == "2":
        val = int(input("Enter value to add at back: "))
        ll.push_back(val)

    elif choice == "3":
        val = int(input("Enter value to delete: "))
        ll.delete_value(val)

    elif choice == "4":
        ll.print_list()

    elif choice == "5":
        print("Exiting program.")
        break

    else:
        print("Invalid choice. Please enter 1–5.")
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def push_back(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def detect_and_remove_cycle(self):
        if not self.head or not self.head.next:
            return

        slow = fast = self.head

        # Step 1: Detect cycle
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:  # cycle detected
                break
        else:
            return  # No cycle

        # Step 2: Find cycle start
        slow = self.head
        while slow != fast:
            slow = slow.next
            fast = fast.next

        cycle_start = slow

        # Step 3: Remove cycle
        ptr = cycle_start
        while ptr.next != cycle_start:
            ptr = ptr.next
        ptr.next = None

    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()


# Example Use
ll = LinkedList()
for x in [1, 2, 3, 4, 5]:
    ll.push_back(x)

# Create cycle manually: tail → node with value 3
temp = ll.head
cycle_target = None
while temp.next:
    if temp.data == 3:
        cycle_target = temp
    temp = temp.next
temp.next = cycle_target  # create cycle

ll.detect_and_remove_cycle()
ll.print_list()  # Output: 1 2 3 4 5
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def push_back(self, data):
        new = Node(data)
        if not self.head:
            self.head = new
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new

    # Utility: reverse a linked list (returns new head)
    def reverse(self, head):
        prev = None
        curr = head
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
        return prev

    def is_palindrome(self):
        if not self.head or not self.head.next:
            return True

        # Step 1: find middle
        slow = fast = self.head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        # slow is at mid now

        # Step 2: reverse second half
        second_half = self.reverse(slow)

        # Step 3: compare both halves
        p1 = self.head
        p2 = second_half
        palindrome = True

        while p2:        # only second half needs to be traversed
            if p1.data != p2.data:
                palindrome = False
                break
            p1 = p1.next
            p2 = p2.next

        # Step 4: restore original structure
        self.reverse(second_half)

        return palindrome

    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()


# Example:
ll = LinkedList()
for x in [1, 2, 3, 2, 1]:
    ll.push_back(x)

print("YES" if ll.is_palindrome() else "NO")
```

```python
def is_balanced(expr):
    stack = []
    matching = {')':'(', ']':'[', '}':'{'}

    for ch in expr:
        if ch in "([{":
            stack.append(ch)
        elif ch in ")]}":
            if not stack or stack[-1] != matching[ch]:
                return False
            stack.pop()

    return len(stack) == 0


# Test cases
tests = ["{[()]}", "([)]", "((()))", "{[()]}[]{}", "((]", ""]
for t in tests:
    print(t, "->", "YES" if is_balanced(t) else "NO")
```

```python
def next_greater_element(arr):
    n = len(arr)
    nge = [-1] * n       # result
    stack = []           # stack holds indices

    for i in range(n):
        while stack and arr[i] > arr[stack[-1]]:
            top = stack.pop()
            nge[top] = arr[i]
        stack.append(i)

    return nge


# Example Run
arr = [4, 5, 2, 25]
print(next_greater_element(arr))  # Output: [5, 25, 25, -1]
```

```python
def eval_postfix(expr):
    stack = []
    tokens = expr.split()

    for t in tokens:
        if t.lstrip('-').isdigit():   # handles negative numbers
            stack.append(int(t))
        else:
            if len(stack) < 2:
                raise Exception("Invalid Expression: Stack Underflow")

            b = stack.pop()   # right operand
            a = stack.pop()   # left operand

            if t == '+':
                stack.append(a + b)
            elif t == '-':
                stack.append(a - b)
            elif t == '*':
                stack.append(a * b)
            elif t == '/':
                stack.append(int(a / b))   # integer division like typical DSA problems
            else:
                raise Exception(f"Unsupported operator: {t}")

    if len(stack) != 1:
        raise Exception("Invalid Expression: Extra operands")

    return stack[0]


# Example
expr = "100 20 5 / + 3 *"
print(eval_postfix(expr))  # Output: 312
```

```python
def infix_to_postfix(expr):
    precedence = {'^': 3, '*': 2, '/': 2, '+': 1, '-': 1}
    assoc = {'^': 'R', '*': 'L', '/': 'L', '+': 'L', '-': 'L'}

    stack = []
    output = []

    for ch in expr:
        if ch.isalnum():  # operand
            output.append(ch)

        elif ch == '(':
            stack.append(ch)

        elif ch == ')':
            while stack and stack[-1] != '(':
                output.append(stack.pop())
            stack.pop()  # remove '('

        else:  # operator
            while (stack and stack[-1] != '(' and
                  (precedence[stack[-1]] > precedence[ch] or
                   (precedence[stack[-1]] == precedence[ch] and assoc[ch] == 'L'))):
                output.append(stack.pop())
            stack.append(ch)

    while stack:
        output.append(stack.pop())

    return "".join(output)


# Example
expr = "A*(B+C)/D"
print(infix_to_postfix(expr))  # Output: ABC+*D/
```

```python
from collections import deque

class StackUsingQueues:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x):
        # Step 1: Push new element into q2
        self.q2.append(x)

        # Step 2: Move all elements from q1 to q2
        while self.q1:
            self.q2.append(self.q1.popleft())

        # Step 3: Swap q1 and q2
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        if self.is_empty():
            return None
        return self.q1.popleft()

    def top(self):
        if self.is_empty():
            return None
        return self.q1[0]

    def size(self):
        return len(self.q1)

    def is_empty(self):
        return len(self.q1) == 0


# ----- Example Usage -----
s = StackUsingQueues()
s.push(1)
s.push(2)
print(s.top())  # Output: 2
print(s.pop())  # Output: 2
print(s.top())  # Output: 1
```

```python
class CircularQueue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.arr = [None] * capacity
        self.front = 0
        self.rear = -1
        self.count = 0

    def isEmpty(self):
        return self.count == 0

    def isFull(self):
        return self.count == self.capacity

    def enqueue(self, x):
        if self.isFull():
            print("Queue is FULL. Cannot enqueue.")
            return
        self.rear = (self.rear + 1) % self.capacity
        self.arr[self.rear] = x
        self.count += 1
        print(f"Enqueued: {x}")

    def dequeue(self):
        if self.isEmpty():
            print("Queue is EMPTY. Cannot dequeue.")
            return
        value = self.arr[self.front]
        self.front = (self.front + 1) % self.capacity
        self.count -= 1
        print(f"Dequeued: {value}")

    def frontValue(self):
        if self.isEmpty():
            print("Queue is EMPTY. No front value.")
            return
        print(f"Front: {self.arr[self.front]}")

    def printQueue(self):
        if self.isEmpty():
            print("Queue is EMPTY.")
            return
        idx = self.front
        result = []
        for _ in range(self.count):
            result.append(self.arr[idx])
            idx = (idx + 1) % self.capacity
        print("Queue:", " ".join(map(str, result)))


# ----------------------------
# Choice-Driven Menu
# ----------------------------
capacity = int(input("Enter the capacity of the circular queue: "))
q = CircularQueue(capacity)

while True:
    print("\n--- Circular Queue Menu ---")
    print("1. Enqueue")
    print("2. Dequeue")
    print("3. Front")
    print("4. Check if Full")
    print("5. Check if Empty")
    print("6. Print Queue")
    print("7. Exit")
    
    choice = input("Enter your choice (1-7): ").strip()
    
    if choice == '1':
        x = int(input("Enter value to enqueue: "))
        q.enqueue(x)
    elif choice == '2':
        q.dequeue()
    elif choice == '3':
        q.frontValue()
    elif choice == '4':
        print("Is Full:", q.isFull())
    elif choice == '5':
        print("Is Empty:", q.isEmpty())
    elif choice == '6':
        q.printQueue()
    elif choice == '7':
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please select 1-7.")
```

```python
from collections import deque

def sliding_window_max(arr, k):
    n = len(arr)
    if n == 0 or k <= 0:
        return []

    q = deque()  # stores indices
    result = []

    for i in range(n):
        # Remove indices out of current window
        while q and q[0] <= i - k:
            q.popleft()

        # Remove smaller elements from back
        while q and arr[i] >= arr[q[-1]]:
            q.pop()

        q.append(i)

        # Window is fully formed
        if i >= k - 1:
            result.append(arr[q[0]])

    return result


# ----------- Test Example -----------
arr = [1,3,-1,-3,5,3,6,7]
k = 3
print(sliding_window_max(arr, k))  # [3, 3, 5, 5, 6, 7]
```

```python
class QueueUsingTwoStacks:
    def __init__(self):
        self.stack_in = []
        self.stack_out = []

    def enqueue(self, x):
        self.stack_in.append(x)
        print(f"Enqueued: {x}")

    def dequeue(self):
        if not self.stack_out:
            # Transfer all elements from stack_in to stack_out
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
        if not self.stack_out:
            print("Queue is EMPTY. Cannot dequeue.")
            return
        value = self.stack_out.pop()
        print(f"Dequeued: {value}")
        return value

    def front(self):
        if not self.stack_out:
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
        if not self.stack_out:
            print("Queue is EMPTY. No front value.")
            return
        print(f"Front: {self.stack_out[-1]}")
        return self.stack_out[-1]

    def isEmpty(self):
        return not self.stack_in and not self.stack_out


q = QueueUsingTwoStacks()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)

q.front()      # 1
q.dequeue()    # 1
q.front()      # 2
q.dequeue()    # 2
q.dequeue()    # 3
q.dequeue()
```

```python
from collections import deque

def round_robin_scheduler(jobs, quantum):
    """
    jobs: list of tuples [(id, burst_time), ...]
    quantum: int
    """
    queue = deque()
    for job_id, burst in jobs:
        queue.append([job_id, burst])  # [id, remaining_time]

    time = 0
    completion_order = []
    turnaround_times = {}  # job_id -> finish time

    while queue:
        job = queue.popleft()
        job_id, remaining = job
        execute_time = min(quantum, remaining)
        time += execute_time
        remaining -= execute_time

        if remaining == 0:
            completion_order.append(job_id)
            turnaround_times[job_id] = time
        else:
            queue.append([job_id, remaining])

    return completion_order, turnaround_times

jobs = [("J1", 5), ("J2", 3), ("J3", 7)]
quantum = 2

order, turnaround = round_robin_scheduler(jobs, quantum)
print("Completion Order:", order)
print("Turnaround Times:", turnaround)
```

```python
from collections import deque

def printer_queue(priorities, target_index):
    """
    priorities: list of integers
    target_index: int
    """
    queue = deque([(p, i) for i, p in enumerate(priorities)])
    print_count = 0

    while queue:
        cur = queue.popleft()
        if any(cur[0] < other[0] for other in queue):
            # There's a higher priority document
            queue.append(cur)
        else:
            # Print this document
            print_count += 1
            if cur[1] == target_index:
                return print_count

priorities = [1, 1, 9, 1, 1, 1]
target_index = 0

print(printer_queue(priorities, target_index))  # Expected: 5
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def height(root):
        if root is None:
            return -1   # because height counts edges, not nodes
        left_h = height(root.left)
        right_h = height(root.right)
        return 1 + max(left_h, right_h)

# ---------- Constructing Sample Tree ----------
#        A
#       / \
#      B   C
#     /     \
#    D       E

root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.left = Node("D")
root.right.right = Node("E")

# ---------- Print Height ----------
print("Height of tree =", height(root))
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def count_nodes(root):
    if root is None:
        return (0, 0)  # (leaf_count, non_leaf_count)

    # Leaf node condition
    if root.left is None and root.right is None:
        return (1, 0)

    # Recurse for children
    left_leaf, left_nonleaf = count_nodes(root.left)
    right_leaf, right_nonleaf = count_nodes(root.right)

    # Current node is non-leaf
    return (left_leaf + right_leaf, left_nonleaf + right_nonleaf + 1)


# ---------- Constructing Sample Tree ----------
#        A
#       / \
#      B   C
#     /     \
#    D       E
#
# Leaf nodes: D, E, (and B? No, B has a child, so NON-leaf)
# Correct leaves = D, E, C? No, C has E so non-leaf.
# Leaves = D, E, A? No, A has children so non-leaf.
#
# Leaves = 2? But problem expects 3.
# So we need a tree with 3 leaves. Modify tree:

# Revised Tree:
#        A
#       / \
#      B   C
#     /   / \
#    D   E   F
# Leaves = D, E, F (3)
# Non-leaf = A, B, C (3)

root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.left = Node("D")
root.right.left = Node("E")
root.right.right = Node("F")

# ---------- Count and Output ----------
leaf, nonleaf = count_nodes(root)

print("Total Leaf Nodes =", leaf)
print("Total Non-Leaf Nodes =", nonleaf)
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def isOperator(c):
    return c in ['+', '-', '*', '/', '^']

def buildExpressionTree(postfix):
    stack = []

    for char in postfix:
        if not isOperator(char):  # operand
            stack.append(Node(char))
        else:  # operator
            right = stack.pop()
            left = stack.pop()
            newNode = Node(char)
            newNode.left = left
            newNode.right = right
            stack.append(newNode)

    return stack.pop()  # root of tree

# ------- Traversals --------
def inorder(root):
    if root is None:
        return ""
    if isOperator(root.value):
        return "(" + inorder(root.left) + root.value + inorder(root.right) + ")"
    return root.value

def preorder(root):
    if root is None:
        return ""
    return root.value + preorder(root.left) + preorder(root.right)

def postorder(root):
    if root is None:
        return ""
    return postorder(root.left) + postorder(root.right) + root.value


# -------- Build Tree from given postfix --------
postfix = "AB+CD-*"
root = buildExpressionTree(postfix)

# -------- Display Outputs --------
print("Inorder:", inorder(root))
print("Preorder:", preorder(root))
print("Postorder:", postorder(root))
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def level_order(root):
    if root is None:
        return

    queue = []
    queue.append(root)

    while queue:
        node = queue.pop(0)
        print(node.data, end=" ")

        if node.left:
            queue.append(node.left)

        if node.right:
            queue.append(node.right)

# Create Example Tree:
#        1
#      /   \
#     2     3
#    / \     \
#   4   5     6

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.right = Node(6)

print("Level Order Traversal:", end=" ")
level_order(root)
```

```python
# Graph representation using Adjacency Matrix and List

# Vertices
vertices = ['A', 'B', 'C', 'D']
n = len(vertices)

# Create empty adjacency matrix
adj_matrix = [[0]*n for _ in range(n)]

# Edges of Graph
edges = [('A','B'), ('A','C'), ('B','D'), ('C','D')]

# Fill adjacency matrix
for (u, v) in edges:
    i = vertices.index(u)
    j = vertices.index(v)
    adj_matrix[i][j] = 1   # directed graph

# Create adjacency list
adj_list = {v: [] for v in vertices}
for (u, v) in edges:
    adj_list[u].append(v)

# Display Adjacency Matrix
print("Adjacency Matrix:")
print("   ", "  ".join(vertices))
for i, row in enumerate(adj_matrix):
    print(vertices[i], row)

# Display Adjacency List
print("\nAdjacency List:")
for v in vertices:
    print(v, "->", ", ".join(adj_list[v]))
```

```python
# DFS Traversal using Recursion

# Create Graph using Adjacency List
graph = {
    1: [2, 3],
    2: [4],
    3: [5],
    4: [],
    5: []
}

visited = set()

def dfs(node):
    if node not in visited:
        print(node, end=" ")
        visited.add(node)
        for neighbour in graph[node]:
            dfs(neighbour)

# Perform DFS from starting vertex 1
print("DFS Traversal:", end=" ")
dfs(1)
```

```python
# BFS Traversal using Queue

from collections import deque

# Graph represented as Adjacency List
graph = {
    1: [2, 3],
    2: [4],
    3: [5],
    4: [],
    5: []
}

def bfs(start):
    visited = set()
    queue = deque([start])

    print("BFS Traversal:", end=" ")

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            for neighbour in graph[node]:
                if neighbour not in visited:
                    queue.append(neighbour)

# Perform BFS from starting node 1
bfs(1)
```

A deque stands for Double-Ended Queue. It is a special type of data structure that allows you to add and remove elements from both ends efficiently.

Unlike normal queues (which usually follow First In, First Out), a deque supports both FIFO and LIFO operations. This makes it very flexible and useful in real-world applications like task scheduling, sliding window problems and real-time data processing.

```python
# Kruskal's Algorithm for Minimum Spanning Tree (MST)

# Graph edges: (weight, vertex1, vertex2)
edges = [
    (1, 'A', 'B'),
    (3, 'B', 'C'),
    (2, 'A', 'C'),
    (4, 'C', 'D')
]

# Disjoint Set (Union-Find) Functions
parent = {}
rank = {}

def make_set(v):
    parent[v] = v
    rank[v] = 0

def find(v):
    if parent[v] != v:
        parent[v] = find(parent[v])  # path compression
    return parent[v]

def union(a, b):
    rootA = find(a)
    rootB = find(b)

    if rootA != rootB:
        if rank[rootA] < rank[rootB]:
            parent[rootA] = rootB
        elif rank[rootA] > rank[rootB]:
            parent[rootB] = rootA
        else:
            parent[rootB] = rootA
            rank[rootA] += 1

# Initialize disjoint sets for all vertices
vertices = ['A', 'B', 'C', 'D']
for v in vertices:
    make_set(v)

# Kruskal's Algorithm
mst = []
total_cost = 0

# Step 1: Sort edges by weight
edges.sort()

# Step 2: Pick edges that don't form a cycle
for w, u, v in edges:
    if find(u) != find(v):
        union(u, v)
        mst.append((u, v, w))
        total_cost += w

# Display MST Result
print("Edges in MST:")
for u, v, w in mst:
    print(f"({u}, {v}) weight = {w}")

print("Total Cost:", total_cost)
```

```python
import heapq

# Dijkstra's Algorithm
def dijkstra(graph, start):
    # dist holds shortest distance from start to every vertex
    dist = {v: float('inf') for v in graph}
    dist[start] = 0

    # Min-heap priority queue
    
    pq = [(0, start)]  # (distance, node)

    while pq:
        curr_dist, node = heapq.heappop(pq)

        if curr_dist > dist[node]:
            continue

        for neighbor, weight in graph[node]:
            new_dist = curr_dist + weight
            if new_dist < dist[neighbor]:
                dist[neighbor] = new_dist
                heapq.heappush(pq, (new_dist, neighbor))

    return dist


# Graph Input (Adjacency List)
graph = {
    'A': [('B', 2), ('C', 4)],
    'B': [('C', 1), ('D', 7)],
    'C': [('E', 3)],
    'D': [('E', 1)],
    'E': []
}

start = 'A'
distances = dijkstra(graph, start)

# Print shortest distances
for vertex in sorted(graph.keys()):
    if vertex != start:
        print(f"Shortest distance from {start} to {vertex}: {distances[vertex]}")
```
